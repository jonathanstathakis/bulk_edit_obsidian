Add the tag "spanish" to all notes in the path.

```python
from pyomd import Note, Notes, metadata
from pathlib import Path

path = "/Users/jonathan/001_obsidian_vault/software"

notes = Notes(Path(path))
[
    print(f"---------- {note.path.name} -------------\n\n", note.content, "\n\n")
    for note in notes.notes
]
```

```python
# add the tag

notes.metadata.add("tags", "software", metadata.MetadataType.FRONTMATTER)
notes.update_content(write=False)
[
    print(f"---------- {note.path.name} -------------\n\n", note.content, "\n\n")
    for note in notes.notes
]
notes.write()
```
